1. Open your project in VS Code

Make sure you are inside the cloned repo folder.

2. Open Terminal in VS Code

Shortcut:

Ctrl + `
3. Check current status
git status

This shows what files you changed.

4. Add your changes
git add .

(“.” means add all files)

5. Commit your changes
git commit -m "your message here"

Example:

git commit -m "updated UI and fixed bugs"
6. Pull latest changes (IMPORTANT ⚠️)

Since you got earlier error (fetch first), do this:

git pull origin main --rebase
7. Push to GitHub
git push origin main
💥 If you still get error (conflicts)

Run:

git pull origin main

Fix conflicts manually in files → then:

git add .
git commit -m "resolved conflicts"
git push origin main
⚡ Alternative (VS Code UI method)
Go to Source Control (left panel)
Click + (stage changes)
Type commit message
Click ✅ Commit
Click ... → Push
🚨 Common mistakes
❌ Not doing git pull before push
❌ Wrong branch (check with git branch)
❌ Not inside repo folder
❌ No internet / GitHub login issue
💡 Pro tip (for you as dev)

Since you're working on real projects:

Use this flow always:

git pull origin main --rebase
git add .
git commit -m "message"
git push origin main
